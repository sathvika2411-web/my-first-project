```python
import pandas as pd
data = {
    'Date': ['2023-01-01', '2023-01-01', '2023-01-02'],
    'Product': ['Chocolate', 'Toy', 'Book'],
    'Quantity': [2, 1, 3],
    'Price': [10, 50, 20]
}
df = pd.DataFrame(data)

print("Sales Table:")
print(df)
```

```python
df['Total'] = df['Quantity'] * df['Price']
print("\nSales Table with Total Price:")
print(df)
```

```python
total_sales = df['Total'].sum()
print("\nTotal Sales = ₹", total_sales)
```

```python
top_product = df.groupby('Product')['Total'].sum().sort_values(ascending=False)
print("\nTop-Selling Products:")
print(top_product)
```

```python
import matplotlib.pyplot as plt

top_product.plot(kind='bar', title='Top-Selling Products', color='skyblue')
plt.ylabel('Total Sales (₹)')
plt.xlabel('Product')
plt.show()
```

```python
top_product.plot(
    kind='pie',
    autopct='%1.1f%%',      
    figsize=(6, 6),        
    title='Sales Contribution by Product',
    ylabel=''         
)
plt.show()
```
